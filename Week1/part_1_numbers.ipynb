### Числа

Числа в Python бувають різними:
 - Цілі числа (int): 122, -4, 99999999999, 0o177, 0x9ff, 0b101010
 - Дійсні числа (float): 1.0, 3.14, .5, 4E21, 4.0e21 
 - Комплексні числа (complex): 3 + 4j, 3.0 + 4.0j,  
 - Числа фіксованої точності% decimal.Decimal('0.1')
 - Раціональні числа: fractions.Fraction(3, 4)

### Комплексні числа

```python
print ("|{0}| = {1}".format(complex(2, 3), abs(complex(2, 3))))
print ("{0} * {1} = {2}".format(complex(1, 1), complex(2, 3), complex(1, 1) * complex(2, 3)))
```

```python
(1 + 2j) * (3+4j)
```

### Числа фіксованої точності і раціональні числа

Для решения проблем, связанных с точностью представления простых вещественных чисел введены
вещественные числа с фиксированной точностью и рациональные числа (числа, представленные дробью, то есть парой целых чисел – числителем и знаменателем).

```python
print(1.1/3)
from decimal import Decimal
from decimal import getcontext
getcontext().prec = 3
Decimal('4') / 3 == Decimal('1.33')
```

```python
from decimal import Decimal
from decimal import getcontext
getcontext().prec = 2
print(Decimal('1.10') / 3)
```

```python
print(7/71)
print(7/71*71 == 7)
```

```python
from fractions import Fraction
print(Fraction(7, 71) * 71 == 7)
```
