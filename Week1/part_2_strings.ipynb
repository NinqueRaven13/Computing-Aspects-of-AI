#### Видалення лишніх пробільних символів


```python
name = ' eric '

print(name.lstrip())
print(name.rstrip())
print(name.strip())
```

Так краще:

```python
name = ' eric '

print('-' + name.lstrip() + '-')
print('-' + name.rstrip() + '-')
print('-' + name.strip() + '-')
```

Позбавлятись можна не тільки від пробілів. Тут ',' - аргумент методів *lstrip()*, *rstrip()* і *strip()*

```python
name = ',,eric,,'

print('-' + name.lstrip(',') + '-')
print('-' + name.rstrip(',') + '-')
print('-' + name.strip(',') + '-')
```

### Методи для работи з рядками

Всі методи рядків:

```python
dir(str)
```

**find** - пошук підрядків в рядку. Повертає номер першого входження або -1

```python
"we bought a new house and we are happy".find("we")
```

**rfind** - пошук підрядків в рядку. Повертає номер першого входження або -1

```python
"we bought a new house and we are happy".rfind("we")
```

**index** - пошук підрядків в рядку. Повертає номер першого входження або визиває ValueError

```python
"we bought a new house and we are happy".index("we")
```

```python
"we bought a new house and we are happy".index("land")
```

**rindex** - пошук підрядків в рядку. Повертає номер останього входження або визиває ValueError

```python
"we bought a new house and we are happy".rindex("we")
```

```python
"we bought a new house and we are happy".rindex("land")
```

**replace** - заміна по шаблону

```python
help(str.replace)
```

```python
"we bought a new house".replace("house", "building")
```

```python
"bla bla bla".replace("bla", "wow", 3)
```

**split** - розбиття рядка по раздільнику

```python
"hello#$great#$to#$see#$you".split("#$")
```

- **isdigit** - чи складається рядок з цифр;
- **isalpha** - чи складається рядок з букв;
- **isalnum** - чи складається рядок з цифр або букв;
- **islower** - чи складається рядок з символів в нижньому регістрі;
- **isupper** - чи складається рядок з символів в верхньому регістрі;
- **istitle** - чи починаються слова в рядку з великої букви.

**join** - Збір рядка з списка по раздільнику, застосовується до роздільника

```python
",".join(["Ann", "Leo", "Tiffany"])
```

**format** - форматування рядка

Приклад

```python
# key-value pairs: client_id, (name, age, occupation)
personal_data = {1: ("Ann", 29, "Artist"), 
               2: ("Leo", 54, "Programmer"), 
               3: ("Tiffany", 43, "Shop Assistant")}
for (name, age, occupation) in personal_data.values():
    print("{0} (age {1}) is a {2}".format(name, age, occupation))
```
