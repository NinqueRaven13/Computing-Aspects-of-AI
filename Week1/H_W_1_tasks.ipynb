## Задача I
Дано три цілих числа. Найдіть найбільше з них (програма повинна вивести рівно одне ціле число).

```python
## Ваш код тут
def GreaterFromThree(a,b,c):
    if a > b & a >c:
        max = a
    elif b > c:
        max = b
    else:
        max = c
    return max

print(GreaterFromThree(2,3,11))
```

## Задача J
Дано три натуральних числа a, b, c. Визначте, чи існує трикутник з такими сторонами. Якщо трикутник існує, виведіть "ТАК", а якщо не є трикутником "НІ".

```python
## Ваш код тут
def IsTriangleExist(a,b,c):
    if (a > 0) & (b > 0) & (c > 0):
        if ((a + b) > c) & ((b + c) > a) & ((c + a) >b):
            existence = True
        else:
            existence = False
    else:
        existence = False
    return existence

print(IsTriangleExist(3,4,5))
```
