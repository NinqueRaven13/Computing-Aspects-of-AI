## Змінні типа Bool. Умовний оператор

```python
# Python 2 and 3 compatibility
# pip install future
from __future__ import (absolute_import, division,
                        print_function, unicode_literals)
from builtins import *
```

#### Для керування логікою програми потрібен ще один тип змінної - bool. Змінна типа bool має всього два значення - True і False.

```python
is_connected = True
print(type(is_connected))

a, b = 5, 6
a_greater_b = a > b
print(a_greater_b)
print(type(a_greater_b))

c = bool(1)
print(c)
print(type(c))
```

#### В найпростішому виді умовна інструкція в Пітоні має наступний синтаксис:

if Умова:
   
    Блок инструкцій 1

else:
   
    Блок инструкцій 2

#### Приклад 
Якщо число ділится на 7, надрукувати це число і рядок "YES", в іншому випадку - "NO"

```python
a = int(input("Insert a number: "))
if a % 7 == 0:
    print(a)
    print("YES")
else:
    print("NO")
```

#### Якщо в кожному блоці всього по одній інструкції, можна використовувати скорочений синтаксис:

Інструкція 1 if Умова else Інструкція 2

```python
has_debts = True
comission = 0.07 if has_debts else 0.05
print(comission)
```

#### Приклад
Визначити, в якій чверті находиться точка, по її координатам *(x, y)*

```python
x = int(input("Введіть x: "))
y = int(input("Введіть y: "))
if x > 0:
    if y > 0:               # x>0, y>0
        print("Перша чверть")
    else:                   # x>0, y<0
        print("Четверта чверть")
else:
    if y > 0:               # x<0, y>0
        print("Друга чверть")
    else:                   # x<0, y<0
        print("Третя чверть")
```

#### Але, не потрібно зловживати вкладеними конструкціями.

#### Блок *if-elif-else* (каскадна конструкція)
if Умова 1:
   
    Блок інструкцій 1

elif Умова 2:

    Блок інструкцій 2
    
...

elif Умова n:

    Блок інструкцій n
    
else:
   
    Блок інструкцій n+1

```python
x = input("Введіть x: ")
y = input("Введіть y: ")
if x > 0 and y > 0:
    print("Перша чверть")
elif x > 0 and y < 0:
    print("Четверта чверть")
elif y > 0:
    print("Друга чверть")
else:
    print("Третя чверть")
```
