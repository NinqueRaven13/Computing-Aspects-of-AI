**Подивимось на Seaborn зразу в дії на даних по моделях місяця по версиї журнала Playboy.**

```python
girls = pd.read_csv('C:/Users/ezsagan/Desktop/DS_пара/python_lesson4/data/girls.csv')
girls.head(10)
```

```python
girls.describe(include='all')
```

**Гістограми. Метод distplot <a href="https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.distplot.html">distplot</a>**

```python
girls['Waist'].hist(bins=15);
```

```python
sns.distplot(girls['Waist'], kde=True);
```

```python
ax = sns.distplot(girls['Height'], kde=False)
ax.set(xlabel='Playboy girls height', ylabel='Frequency')
sns.set_style('darkgrid')
```

**Метод boxplot <a href="https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.boxplot.html">boxplot</a>**
В наборі даних всі признаки чисельні, так що створимо категорію "weight_cat" з 3 типів ваги.

```python
def weight_category(weight):
    return 'heavier' if weight > 54\
            else 'lighter' if weight < 49 else 'median'

girls['weight_cat'] = girls['Weight'].apply(weight_category)
sns.boxplot(x='weight_cat', y='Height', data=girls);
```

**Метод pairplot <a href="http://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.pairplot.html">pairplot</a>**

```python
sns.set_palette(sns.color_palette("RdBu"))
sns.pairplot(girls[['Bust', 'Waist', 'Hips', 'Height', 'Weight']]);
```

```python
girls.corr()
```

```python
girls.head()
```

**Метод countplot <a href="http://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.countplot.html">countplot</a>**

```python
def height_category(height):
    return 'high' if height > 175\
            else 'small' if height < 160 else 'median'

girls['height_cat'] = girls['Height'].apply(height_category)
sns.countplot(x='height_cat', hue='weight_cat', data=girls);
```

```python
pd.crosstab(girls['weight_cat'], girls['height_cat'])
```

**Метод jointplot [jointplot](http://seaborn.pydata.org/generated/seaborn.jointplot.html)**

```python
sns.jointplot(x='Weight', y='Height', 
              data=girls, kind='reg');
```

## Приклад візуального аналіза даних з Pandas і Seaborn 

В 1968 році була [опублікована](http://www.ncbi.nlm.nih.gov/pubmed/5676802) стаття під назвою "Correlation of Performance Test Scores with Tissue Concentration of Lysergic Acid Diethylamide in Human Subjects".

До статті прикріплен невеликий набір даних всього з 7 спостережень.

```python
data_types = {'Drugs': float,
              'Score': float}
df = pd.read_csv('../../data/drugs-and-math.csv', 
                 index_col=0, sep=',', dtype=data_types)
```

```python
print(df.shape)
print(df.columns)
print(df.index)
```

**Таблиця вже відсортована по колонці Drugs, зробимо сортировку по Score.**

```python
df.sort_values('Score', 
               ascending=False, 
               inplace=True)
```
